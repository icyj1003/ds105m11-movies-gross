# Chi tiết từng thuộc tính

## Genre: Thể loại

```python
def getList(feature):
    gs = df[feature]

    gs.dropna(inplace=True)

    gList = ''

    for g in gs:
        gList = gList + g

    gList = list(set(gList.split(',')))
    try:
        gList.remove('')
    except:
        pass
    return gList
```

Tất cả thể loại

```python
genreList = getList('genres')
print(len(genreList), 'Gồm thể loại:')
for stt, genre in zip(range(0, len(genreList)), genreList):
    if stt+1 < 10:
        print(f"{stt+1}  {genre}")
    else:
        print(f"{stt+1} {genre}")
```

Số lượng phim theo thể loại

```python
dataG = []
for genre in genreList:
    for line in df[df.genres.str.contains(genre, na=False)].values:
        dataG.append((genre, line[6]))
dataG = pd.DataFrame(data=dataG, columns=['genre', 'gross'])
dataG.genre.value_counts()
```

```python
fig, ax = plt.subplots()
plt.ticklabel_format(style='plain', axis='y')
fig.set_size_inches(len(genreList) * 5, 20)
fig.set_dpi(200)
ax.xaxis.set_tick_params(labelsize='large')
sb.boxplot(x='genre', y='gross', data=dataG, ax=ax, showfliers=False)
```

## Ngôn ngữ

```python
langList = getList('languages')

dataL = []
for lang in langList:
    for line in df[df.languages.str.contains(lang, na=False)].values:
        dataL.append((lang, line[6]))
dataL = pd.DataFrame(data=dataL, columns=['lang', 'gross'])
dataL.lang.value_counts().to_csv('l.csv')
```
