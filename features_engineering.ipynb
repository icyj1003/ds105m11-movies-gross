```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import pandas as pd
import numpy as np
import seaborn as sb
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('./data/data_preprocessed.csv')
```

```python
df.head(5)
```

```python
genres = df.genres

df.languages.value_counts().to_csv('ll.csv')

genres.dropna(inplace=True)

genreList = ''

for genre in genres:
    genreList = genreList + genre

genreList = list(set(genreList.split(',')))

genreList.remove('')

G = []

for row in df.genres:
    rG = []
    for genre in genreList:
        if str(row).find(genre)!=-1:
            rG.append(1)
        else:
            rG.append(0)

    G.append(rG)

dfG = pd.DataFrame(G)

dfG.columns = genreList

df = pd.concat([df, dfG], axis=1)

df.head(5)
```

```python
languages = df.languages

languages.dropna(inplace=True)

languageList = ''

for language in languages:
    languageList = languageList + language

languageList = list(set(languageList.split(',')))

languageList.remove('')

G = []

for row in df.languages:
    rG = []
    for language in languageList:
        if str(row).find(language)!=-1:
            rG.append(1)
        else:
            rG.append(0)

    G.append(rG)

dfG = pd.DataFrame(G)

dfG.columns = languageList

df = pd.concat([df, dfG], axis=1)

df.head(5)
```

```python
languages = df.countries_of_origin

languages.dropna(inplace=True)

languageList = ''

for language in languages:
    languageList = languageList + language

languageList = list(set(languageList.split(',')))

languageList.remove('')

len(languageList)
```
